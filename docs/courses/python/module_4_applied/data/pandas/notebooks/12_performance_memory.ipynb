```python
optimized = orders.copy()
optimized["category"] = optimized["category"].astype("category")
optimized["shipping_status"] = optimized["shipping_status"].astype("category")

optimized["quantity"] = pd.to_numeric(optimized["quantity"], downcast="integer")
optimized["price"] = pd.to_numeric(optimized["price"], downcast="float")

optimized.memory_usage(deep=True).sum(), optimized.dtypes
```
