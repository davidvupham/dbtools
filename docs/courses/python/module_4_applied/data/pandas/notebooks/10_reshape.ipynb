```python
pivot = orders.pivot_table(index="customer", columns="category", values="revenue", aggfunc="sum", fill_value=0)
pivot
```

```python
wide = pivot.reset_index()
long = wide.melt(id_vars=["customer"], var_name="category", value_name="revenue")
long.head()
```

```python
stacked = pivot.stack()
stacked.head()
```
