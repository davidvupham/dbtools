```python
import pandas as pd

# Example reads (requires local files):
# df = pd.read_csv("orders.csv")
# df = pd.read_parquet("data.parquet")

# Demonstrate dtype inspection pattern with a tiny in-memory df

df = pd.DataFrame({"order_id": [1, 2], "order_date": ["2025-01-01", "2025-01-02"], "price": ["9.99", "10.50"]})
df.dtypes
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df.dtypes
```
